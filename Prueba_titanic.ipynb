```python
import pandas as pd #biblioteca para el análisis y la manipulación de datos
```

```python
dataset = pd.read_csv("titanic.csv") #leer datasets
```

```python
dataset.head(10) #mostrar primeros 10 del dataset
```

```python
dataset.tail(10) #mostrar 10 últimos
```

```python
dataset.describe() #genera un resumen estadístico de las columnas numéricas de un DataFrame o Series
```

```python
dataset.info() #Muestra informacion sobre el dataset
```

```python
valores_faltantes = dataset.isna().sum() #valores no asignados y la cantidad
print(valores_faltantes)
```

```python
columnas_con_na = dataset.columns[dataset.isna().any()]
print(columnas_con_na)
```

```python
dataset.Embarked.value_counts() #Cuenta la cantidad de ocurrencias únicas de cada valor en la columna Embarked
```

```python
#NA
dataset.Age.fillna(value=dataset.Age.mean()) # Establece en los valores NA la media de valores de la columna 'Age'
valores_faltantes = dataset.isna().sum()
print(valores_faltantes)
```

```python
dataset.Embarked.fillna(value=(dataset.Embarked.value_counts().idxmax())) #Llena los NA con el índice de mayor valor de la columna ‘Embarked’
columnas_con_na = dataset.columns[dataset.isna().any()]
print(columnas_con_na)
```

```python
dataset.Embarked.value_counts() #Ahora hay dos 'S' más que antes
```

```python
fsiz = dataset.apply(lambda x: x.SibSp+x.Parch, axis=1) #Crear una Serie con la suma de SibSp y Parch
dataset['Family_Size'] = fsiz # Agrego la serie al DataFrame 'dataset' con el nombre de columna 'Family_Size'
dataset.head(3)
```

```python
dataset.drop('Parch', axis=1, inplace=True) #Eliminar columna 'Parch'
dataset.drop('SibSp', axis=1, inplace=True)
dataset.drop('Name', axis=1, inplace=True)
dataset.drop('Cabin', axis=1, inplace=True)
dataset.drop('Ticket', axis=1, inplace=True)
dataset.drop('PassengerId', axis=1, inplace=True)
dataset.head(3)
```

```python
dataset.dtypes #muestra el tipo de dato de cada columna
```

```python
for col in dataset.select_dtypes(exclude=["number"]).columns:
    print("Converting column "+col+"...")
    dataset[col] = dataset[col].astype('category')  #Convierte el tipo de dato en categórico
    print(dataset[col].cat.categories) #Este atributo devuelve las categorías únicas de la columna categórica.
    dataset[col] = dataset[col].cat.codes  #convierte las categorías en códigos numéricos.
```

```python
dataset.dtypes
```

```python
dataset.tail(3)
```

```python
# Separar datos
Y = dataset['Survived'] #Almacenamos en 'Y' la columna 'Survived'
X = dataset.drop('Survived', axis=1) #Almacenamos en 'X' las filas del dataset - la columna 'Survived'
```

```python
print(X.shape, Y.shape)
```

```python
# SCIKIT LEARN
import sklearn
```

```python
# training y testing
from sklearn.model_selection import train_test_split
```

```python
TRAIN_SIZE = 0.8
TESTING_SIZE = 0.2
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=TESTING_SIZE, random_state=42) #Separación para training y testing
```

```python
print(x_train.shape, y_train.shape)
```

```python
print(x_test.shape, y_test.shape)
```

```python
# ARBOL DECISIONES  - DECISION TREE
from sklearn.tree import DecisionTreeClassifier
```

```python
modelo_1 = DecisionTreeClassifier(random_state=0) #Crear arbol de decisiones
```

```python
modelo_1.fit(x_train, y_train) #Entrenar el modelo
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
modelo_2 = RandomForestClassifier(n_estimators=15) # Crear Random Forest
```

```python
modelo_2.fit(x_train, y_train) #Entrenar modelo2
```

```python
print(modelo_1.score(x_train, y_train))
```

```python
print(modelo_2.score(x_train, y_train))
```

```python
# EVALUACION
#Testing dataset
print(modelo_1.score(x_test, y_test))
```

```python
print(modelo_2.score(x_test, y_test))
```
