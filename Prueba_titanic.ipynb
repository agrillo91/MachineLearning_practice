```python
import pandas as pd #biblioteca para el análisis y la manipulación de datos
```

```python
dataset = pd.read_csv("titanic.csv") #leer datasets
```

```python
dataset.head(10) #mostrar primeros 10 del dataset
```

```python
dataset.tail(10) #mostrar 10 últimos
```

```python
dataset.describe() #genera un resumen estadístico de las columnas numéricas de un DataFrame o Series
```

```python
dataset.info() #Muestra informacion sobre el dataset
```

```python
valores_faltantes = dataset.isna().sum() #valores no asignados y la cantidad
print(valores_faltantes)
```

```python
columnas_con_na = dataset.columns[dataset.isna().any()]
print(columnas_con_na)
```

```python
dataset.Embarked.value_counts() #Cuenta la cantidad de ocurrencias únicas de cada valor en la columna Embarked
```

```python
#NA
dataset.Age.fillna(value=dataset.Age.mean()) # Establece en los valores NA la media de valores de la columna 'Age'
valores_faltantes = dataset.isna().sum()
print(valores_faltantes)
```

```python
dataset.Embarked.fillna(value=(dataset.Embarked.value_counts().idxmax())) #Llena los NA con el índice de mayor valor de la columna ‘Embarked’
columnas_con_na = dataset.columns[dataset.isna().any()]
print(columnas_con_na)
```

```python
dataset.Embarked.value_counts() #Ahora hay dos 'S' más que antes
```

```python
fsiz = dataset.apply(lambda x: x.SibSp+x.Parch, axis=1) #Crear una Serie con la suma de SibSp y Parch
dataset['Family_Size'] = fsiz # Agrego la serie al DataFrame 'dataset' con el nombre de columna 'Family_Size'
dataset.head(3)
```

```python
dataset.drop('Parch', axis=1, inplace=True) #Eliminar columna 'Parch'
dataset.drop('SibSp', axis=1, inplace=True)
dataset.drop('Name', axis=1, inplace=True)
dataset.drop('Cabin', axis=1, inplace=True)
dataset.drop('Ticket', axis=1, inplace=True)
dataset.drop('PassengerId', axis=1, inplace=True)
dataset.head(3)
```
